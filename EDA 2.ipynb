# collaborative filtering based recommender system

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
mean_votes = np.mean(df['IMDb Votes']).round(0)
mean_votes = mean_votes.astype(int)
mean_votes
```

```python
import math
df['IMDb Votes'] = df['IMDb Votes'].apply(lambda x: mean_votes if x==0 or math.isnan(x) else x)
df['IMDb Votes'] = df['IMDb Votes'].astype(int)
df['IMDb Votes']
```

```python
df.info()
```

```python
df.rename(columns={'View Rating': 'Viewersrating'}, inplace=True)
df.columns
```

```python
df['Actors'] = df['Actors'].astype(str)
df['Viewersrating'] = df['Viewersrating'].astype(str)
df['Summary'] = df['Summary'].astype(str)
df['Genre'] = df['Genre'].astype(str)
df['Title'] = df['Title'].astype(str)
```

```python
def clean_data(x):
    return str.lower(x.replace(" ", ""))
```

```python
new_features = ['Title','Genre', 'Summary', 'Actors', 'Viewersrating']
selected_data = df[new_features]
selected_data
```

```python
for new_feature in new_features:
    selected_data.loc[:, new_feature] = selected_data.loc[:, new_feature].apply(clean_data)
selected_data
```

```python
def soup(x):
    return x['Genre'] + ' ' + x['Summary'] + ' ' + x['Viewersrating']
```

```python
selected_data['soup'] = soup(selected_data)
selected_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(stop_words='english')
count_matrix = cv.fit_transform(selected_data['soup'])
cosine_sim2 = cosine_similarity(count_matrix)
cosine_sim2
```

```python
selected_data.reset_index(inplace=True)
selected_data.head()
```

```python
indices = pd.Series(selected_data.index, index=selected_data['Title'])
indices
```

```python
cosine_sim2
```

```python
def recommend(title, indices=indices, cosine_sim= cosine_sim2, num_recommendations=5):
    title = title.replace(' ', '').lower()
    if title not in indices:
        print("Sorry :( We don't have this movie in the dataset")
    idx = indices[title]
    sim_scores = cosine_sim[idx]
    ranked_indices = np.argsort(sim_scores)[::-1]
    closest_indices = ranked_indices[1:num_recommendations + 1]
    closest_movies = [(df.iloc[i]['Title']) for i in closest_indices]

    return closest_movies
recommend('Breaking Bad')
```

```python
i = indices['breakingbad']            # this fetches the movie indice 
print(cosine_sim2[i]) # this find the similarity vectors near the movie above
ri = np.argsort(cosine_sim2[i])[::-1] # this sorts the movies based their cosine similarity with the given movie input
print(ri)
recommending=5# we will recommend top 5 shows
near = ri[1: recommending + 1] # recommending the top 5 from the ri(ranked indices)
print(near)
# now we need to get the title of the show on that indice using list comprehension and accessing the "near" variable
ti = [(df.iloc[i] ['Title']) for i in near]
print(ti)
```
